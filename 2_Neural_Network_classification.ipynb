### Create a function `plot_decision_boundary()`

* Take in a trained model, features (x) and labels(y)
* Create a meshgrid of the different x values
* Make predictions as well as a line b/w zones (where each unique class falls)

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.c_[np.array([1,2,3]), np.array([4,5,6])]
```

```python
def plot_decision_boundary(model, x, y):
  x_min, x_max = x[:,0].min() - 0.1, x[:,0].max()+0.1
  y_min, y_max = x[:,1].min() - 0.1, x[:,1].max()+0.1
  xx, yy = np.meshgrid(np.linspace(x_min, x_max, 100),
                       np.linspace(y_min, y_max, 100))

  x_in = np.c_[xx.ravel(), yy.ravel()]

  y_pred = model.predict(x_in)

  if len(y_pred[0]) > 1:
    print("doing multiclass classification")
    y_pred = np.argmax(y_pred, axis=1).reshape(xx.shape)

  else:
    print("doing binary classification")
    y_pred = np.round(y_pred).reshape(xx.shape)

  plt.contour(xx, yy, y_pred, cmap=plt.cm.RdYlBu, alpha=0.7)
  plt.scatter(x[:,0], x[:,1], c=y, s=40, cmap=plt.cm.RdYlBu )
  plt.xlim(xx.min(), xx.max())
  plt.ylim(yy.min(), yy.max())
```

```python
plot_decision_boundary(model = model_3,
                       x = x,
                       y = y);
```

### Finding the best learning rate

Use the following Steps :
1. A Learning **Callback**  
2. Another model
3. Modified loss curve plot

```python
tf.random.set_seed(42)

model_8 = tf.keras.Sequential([
    tf.keras.layers.Dense(4, activation="relu"),
    tf.keras.layers.Dense(4, activation="relu"),
    tf.keras.layers.Dense(1, activation="sigmoid")
])

model_8.compile(
    loss = "binary_crossentropy",
    optimizer = tf.keras.optimizers.Adam(),
    metrics = ["accuracy"]
)

# Create a learning rate callback
lr_scheduler = tf.keras.callbacks.LearningRateScheduler(lambda epoch: 1e-4 * 10**(epoch/20))    ##  1e-4 = 0.0001

history_8 = model_8.fit(x_train,y_train, epochs = 100, callbacks = [lr_scheduler])
```

```python
pd.DataFrame(history_8.history).plot(figsize=(10,7), xlabel = "epochs")
```

```python
pd.DataFrame(history_8.history).head()
```

```python
lrs = 1e-4 * 10**(tf.range(100)/20)
plt.figure(figsize=(10,7))
plt.semilogx(lrs, history_8.history["loss"])
plt.xlabel("Learning Rate")
plt.ylabel("Loss")
plt.title("Learning Rate vs loss");
```

## Building a multiclass classification model

For our mutli-class classification model, we can use a similar architecture to our binary classifiers, however going to have to tweak a few things
* Input shape = 28 x 28 (per image)
* Output shape = 10
* Loss Function = tf.keras.losses.CategoricalCrossentropy()
* Output layer activation = Softmax (not sigmoid)

```python
flatten_model = tf.keras.Sequential
```

```python
tf.random.set_seed(42)

model_11 = tf.keras.Sequential([
    tf.keras.layers.Flatten(input_shape=(28, 28)),
    tf.keras.layers.Dense(4, activation ="relu"),
    tf.keras.layers.Dense(4, activation="relu"),
    tf.keras.layers.Dense(10, activation="softmax")
])

model_11.compile(
    loss = tf.keras.losses.CategoricalCrossentropy(),
    optimizer = tf.keras.optimizers.Adam(),
    metrics = ["accuracy"]
)

non_history = model_11.fit(train_data, tf.one_hot(train_labels, depth = 10), epochs = 10, validation_data=(test_data, tf.one_hot(test_labels, depth =10)))
```

### Improving the model by **Normalizing the data**

```python
tf.random.set_seed(42)

model_12 = tf.keras.Sequential([
    tf.keras.layers.Flatten(input_shape=(28, 28)),
    tf.keras.layers.Dense(4, activation ="relu"),
    tf.keras.layers.Dense(4, activation="relu"),
    tf.keras.layers.Dense(10, activation="softmax")
])

model_12.compile(
    loss = tf.keras.losses.CategoricalCrossentropy(),
    optimizer = tf.keras.optimizers.Adam(),
    metrics = ["accuracy"]
)

norm_history = model_12.fit(train_data_norm, tf.one_hot(train_labels, depth =10), epochs = 10, validation_data=(test_data_norm, tf.one_hot(test_labels, depth = 10)))
```

```python
tf.random.set_seed(42)

model_13 = tf.keras.Sequential([
    tf.keras.layers.Flatten(input_shape=(28, 28)),
    tf.keras.layers.Dense(4, activation ="relu"),
    tf.keras.layers.Dense(4, activation="relu"),
    tf.keras.layers.Dense(10, activation="softmax")
])

model_13.compile(
    loss = tf.keras.losses.SparseCategoricalCrossentropy(),
    optimizer = tf.keras.optimizers.Adam(),
    metrics = ["accuracy"]
)

lr_scheduler = tf.keras.callbacks.LearningRateScheduler(lambda epoch: 1e-3 * 10**(epoch/20))

find_lr_history = model_13.fit(train_data_norm,train_labels, epochs = 40, validation_data=(test_data_norm, test_labels))
```
