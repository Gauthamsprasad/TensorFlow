### Steps in modelling with Tensorflow

1. **Creating a model** - define the input and output layers, as well as the hidden layers of a deep learning model.

2. **Compiling a model** - define the loss function ( in other words , the function which tells our model how wrong it is ) and the optimizer (tells our model how to improve the patterns its learning) and evaluate matrices (what we can use to interpret the performance of our model).

3. **Fitting a model** - letting the model to find the patters between x and y (features and labels)

````python
# Set random seed
tf.random.set_seed(42)

# 1. Create a model using the Sequential API

model = tf.keras.Sequential([
    tf.keras.layers.Dense(1)
])

# 2. Compile the model
model.compile(loss = tf.keras.losses.mae,
              optimizer=tf.keras.optimizers.SGD(),
              metrics = ["mae"])

# 3. Fit the model
model.fit(x, y, epochs=5)
````

### Improve our model

````python
# Lets rebuild our model
tf.random.set_seed(42)
# 1. Create the model
model = tf.keras.Sequential([
    tf.keras.layers.Dense(1)
])

# 2. Optimize the model

model.compile(
    loss = tf.keras.losses.mae,
    optimizer = tf.keras.optimizers.SGD(),
    metrics=["mae"]
    )

# 3. Fit the model (this time we'll train for longer)

model.fit(x, y, epochs=100)
````

````python
# Let's see the predcition improved
model.predict(tf.constant([[17.0]]))
````

````python
# Let's see if we can make another improvment

# Lets rebuild our model


# 1. Create the model (with extra hidden layers with 100 hidden units)

model = tf.keras.Sequential([
    tf.keras.layers.Dense(50, activation=None),
    tf.keras.layers.Dense(1)
])

# 2. Optimize the model

model.compile(
    loss = tf.keras.losses.mae,
    optimizer = tf.keras.optimizers.Adam(learning_rate=0.01),
    metrics=["mae"])

# 3. Fit the model (this time we'll train for longer)

model.fit(x, y, epochs=100)
````

````python
def mae(y_true, y_pred):
  m =  tf.keras.losses.mae(y_true = y_test,y_pred = tf.constant(y_pred))
  avg = tf.reduce_mean(m)
  return avg.numpy()

def mse(y_true, y_pred):
  m =  tf.keras.losses.mse(y_true = y_test,y_pred = tf.constant(y_pred))
  avg = tf.reduce_mean(m)
  return avg.numpy()
````

````python
import pandas as pd
import matplotlib.pyplot as plt
import tensorflow as tf

#Read insurance dataset
insurance = pd.read_csv("https://raw.githubusercontent.com/stedy/Machine-Learning-with-R-datasets/refs/heads/master/insurance.csv")
insurance
````

Tp prepare the data we need the help of Sciket-learn

````python
from sklearn.compose import make_column_transformer
from sklearn.preprocessing import MinMaxScaler, OneHotEncoder
from sklearn.model_selection import train_test_split

# Create a column transformer
ct = make_column_transformer(
    (MinMaxScaler(), ["age","bmi","children"]), # Turn all values in this cloumns to be 0 - 1 range
    (OneHotEncoder(handle_unknown="ignore"), ["sex","smoker","region"])

)

# create x & y
x = insurance.drop("charges",axis=1)
y = insurance["charges"]

#split the train test data
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

#fit the column transformer to our training data
ct.fit(x_train)

#transform training and test data with normalization (MinmaxScaler) and OneHotEncoder
x_train_normal = ct.transform(x_train)
x_test_noraml = ct.transform(x_test)
````

Beautiful! Our data has been normalized and one hot encoded, now lets build NN model on it

````python
tf.random.set_seed(42)

insurance_model_final = tf.keras.Sequential([
    tf.keras.layers.Dense(100),
    tf.keras.layers.Dense(10),
    tf.keras.layers.Dense(1)
])

insurance_model_final.compile(
    loss = tf.keras.losses.mae,
    optimizer = tf.keras.optimizers.Adam(),
    metrics = ["mae"]
)

insurance_model_final.fit(x_train_normal, y_train, epochs = 200)
````

````python
insurance_model_final.evaluate(x_test_noraml,y_test)
````
